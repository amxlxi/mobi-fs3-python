**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img):
    result = numpy.zeros(img.shape)
    n = 1
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        if p[0] >= n and p[0] < img.shape[0] - n and p[1] >= n and p[1] < img.shape[1] - n:
            for q in numpy.ndindex(3, 3):
                if q[1] == 0:
                    neighborhood_sum -= img[p[0] + q[0] - n,p[1] + q[1] - n]      #bc of flipping, we had to switch + and -
                if q[1] == 2:
                    neighborhood_sum += img[p[0] + q[0] - n,p[1] + q[1] - n]
                
            result[p] = 1/6* (neighborhood_sum)
    return result
        # ...
    # write your code here
    
def prewitt_v(img):
    result = numpy.zeros(img.shape)
    n = 1
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        if p[0] >= n and p[0] < img.shape[0] - n and p[1] >= n and p[1] < img.shape[1] - n:
            for q in numpy.ndindex(3, 3):
                if q[0] == 0:
                    neighborhood_sum -= img[p[0] + q[0] - n,p[1] + q[1] - n]
                if q[0] == 2:
                    neighborhood_sum += img[p[0] + q[0] - n,p[1] + q[1] - n]
                
            result[p] = 1/6* (neighborhood_sum)
    return result
    # write your code here
```
